```python
# Fake news Classification using LSTM

import pandas as pd

df = pd.read_csv("D:/ML/Fake News LSTM/train.csv")
```

```python
df.head
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
X = df.drop('label',axis =1)
```

```python
X.shape
```

```python
y=df['label']
y.shape
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding,Dense,LSTM
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
```

```python
vocab_size =5000
```

```python
messages = X.copy()
```

```python
type(messages)
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
#Preprocessing

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

corpus = []

for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]', ' ' , messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english') ]
    review = ' '.join(review)
    corpus.append(review)
            
                      
```

```python
type(corpus)
```

```python
corpus
```

```python
one_hot_rp = [one_hot(word,vocab_size) for word in corpus]
```

```python
one_hot_rp
```

```python
sent_length =20
embedded_docs = pad_sequences(one_hot_rp, padding = 'pre', maxlen = sent_length)
print(embedded_docs)
```

```python
# Create Model

embedding_feature_vector = 40

model = Sequential()
model.add(Embedding(vocab_size,embedding_feature_vector,input_length = sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

print(model.summary())
```

```python
import numpy as np

X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape
```

```python
y_final.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size=0.33,random_state=42)
```

```python
model.fit(X_train,y_train,validation_data = (X_test,y_test),epochs =10, batch_size=64)
```

```python
y_pred = model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
